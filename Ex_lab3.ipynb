```python
import requests


page = requests.get('https://www.allrecipes.com/gallery/top-recipes-july-2022/')



print(page.content)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(page.content, 'html.parser')
```

```python
a = []
print("here all the paragraph found it:")
for text in soup.find_all('p'):
    b = text.get_text()
    a.append(b)
print(a)
```

```python
a = []
print("here all the matchs found:")
for text in soup.find_all('h2'):
    b = text.get_text()
    a.append(b)
print(a)
```
